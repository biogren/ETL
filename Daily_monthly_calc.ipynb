```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('air_data/air-2022-transformed.csv')

df.info()

```

```python
# Convert the datetime column to a pandas datetime object
df['Date/Time'] = pd.to_datetime(df['Date/Time'])

df.info()
```

```python
columns =["SO2", "NO2", "CO", "O3", "PM10", "PM2.5", "TEMP", "WD10", "WS10"]

# Group the data by month and calculate the average for each month
monthly_data = df.groupby(pd.Grouper(key='Date/Time', freq='M'))[columns].mean()


# Write the monthly data to a new CSV file
monthly_data.to_csv('air_data/monthly_data.csv')
```

```python
monthly_data
```

```python
columns =["SO2", "NO2", "CO", "O3", "PM10", "PM2.5", "TEMP", "WD10", "WS10"]

# Group the data by month and calculate the average for each day
daily_data = df.groupby(pd.Grouper(key='Date/Time', freq='D'))[columns].mean()


# Write the monthly data to a new CSV file
daily_data.to_csv('air_data/daily_data.csv')

daily_data
```
